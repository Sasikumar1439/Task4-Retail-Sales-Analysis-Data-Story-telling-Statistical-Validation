### **Business Hypothesis**

```python
# H0: Discounts do not affect customer spending  
# H1: Discounts significantly affect customer spending
from scipy.stats import ttest_ind

# Example: compare two payment methods
group1 = df[df['payment_method'] == 'Cash']['total_spent']
group2 = df[df['payment_method'] == 'Credit Card']['total_spent']
df.groupby('payment_method')['total_spent'].mean()

print("Group1:", len(group1))
print("Group2:", len(group2))

t_stat, p_value = ttest_ind(group1, group2)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```
